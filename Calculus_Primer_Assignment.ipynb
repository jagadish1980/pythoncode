```python
def gradient_descent(J, dJ, x, alpha, tol, max_iters):
    """
    Performs gradient descent to minimize the objective function J.

    Arguments:
    J -- objective function to be minimized
    dJ -- gradient of the objective function
    x -- initial guess for the solution
    alpha -- learning rate
    tol -- tolerance for convergence
    max_iters -- maximum number of iterations to perform

    Returns:
    x -- solution to the optimization problem
    J_history -- vector of cost function values for each iteration (num_iters x 1)
    """
    J_history = []
    iter = 0

    while iter < max_iters:
        # Compute the gradient
        grad = dJ(x)

        # Update x
        x_new = x - alpha * grad

        # Compute the improvement in the objective function
        improvement = abs(J(x_new) - J(x))

        # Store the cost value
        J_history.append(J(x))

        # Update x and iteration counter
        x = x_new
        iter += 1

        # Check for convergence
        if improvement < tol:
            break

    return x, J_history

# Define the cost function J(x)
def J(x):
    return x**2 - 4*x + 3

# Define the gradient of the cost function dJ(x)
def dJ(x):
    return 2*x - 4

# Parameters for Gradient Descent
initial_x = 0  # Initial guess
alpha = 0.1    # Learning rate
tol = 1e-6     # Tolerance for convergence
max_iters = 1000  # Maximum number of iterations

# Perform Gradient Descent
x_final, J_history = gradient_descent(J, dJ, initial_x, alpha, tol, max_iters)

# Output the results
print("Optimized parameter value:", x_final)
print("Energy costs over steps:", J_history)
```

```python
import matplotlib.pyplot as plt

def J(x):
    return x**2 - 4*x + 3

def dJ(x):
    return 2*x - 4

def gradient_descent(J, dJ, x, alpha, tol, max_iters):
    """
    Performs gradient descent to minimize the objective function J.

    Arguments:
    J -- objective function to be minimized
    dJ -- gradient of the objective function
    x -- initial guess for the solution
    alpha -- learning rate
    tol -- tolerance for convergence
    max_iters -- maximum number of iterations to perform

    Returns:
    x -- solution to the optimization problem
    J_history -- vector of cost function values for each iteration (num_iters x 1)
    """
    J_history = []
    iter = 0

    while iter < max_iters:
        # Compute the gradient
        grad = dJ(x)

        # Update x
        x_new = x - alpha * grad

        # Compute the improvement in the objective function
        improvement = abs(J(x_new) - J(x))

        # Store the cost value
        J_history.append(J(x))

        # Update x and iteration counter
        x = x_new
        iter += 1

        # Check for convergence
        if improvement < tol:
            break

    return x, J_history

# Parameters for Gradient Descent
alpha = 0.1    # Learning rate
tol = 1e-6     # Tolerance for convergence
max_iters = 1000  # Maximum number of iterations

# Different initializations for x
initializations = [-5, -3, 0, 3, 5]

results = []
for init_x in initializations:
    # Perform Gradient Descent
    x_final, J_history = gradient_descent(J, dJ, init_x, alpha, tol, max_iters)

    # Store results: initial x, final x, number of iterations, final cost
    results.append({
        'initial_x': init_x,
        'final_x': x_final,
        'num_iterations': len(J_history),
        'final_cost': J_history[-1],
        'J_history': J_history
    })

# Print results
for result in results:
    print(f"Initialization: {result['initial_x']}")
    print(f"Optimized parameter value: {result['final_x']}")
    print(f"Number of iterations: {result['num_iterations']}")
    print(f"Final cost value: {result['final_cost']}")
    print("-" * 50)

# Select the initialization with the best performance (e.g., fewest iterations or lowest final cost)
best_result = min(results, key=lambda x: x['num_iterations'])
print(f"Best initialization based on convergence time: {best_result['initial_x']}")

# Plotting the cost function values over iterations for each initialization
plt.figure(figsize=(10, 6))
for result in results:
    plt.plot(result['J_history'], label=f'Init x={result["initial_x"]}')

plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Cost Function Value over Iterations for Different Initializations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def J(x):
    return x**2 - 4*x + 3

def dJ(x):
    return 2*x - 4

def gradient_descent(J, dJ, x, alpha, tol, max_iters):
    """
    Performs gradient descent to minimize the objective function J.

    Arguments:
    J -- objective function to be minimized
    dJ -- gradient of the objective function
    x -- initial guess for the solution
    alpha -- learning rate
    tol -- tolerance for convergence
    max_iters -- maximum number of iterations to perform

    Returns:
    x -- solution to the optimization problem
    J_history -- vector of cost function values for each iteration (num_iters x 1)
    """
    J_history = []
    iter = 0

    while iter < max_iters:
        # Compute the gradient
        grad = dJ(x)

        # Update x
        x_new = x - alpha * grad

        # Compute the improvement in the objective function
        improvement = abs(J(x_new) - J(x))

        # Store the cost value
        J_history.append(J(x))

        # Update x and iteration counter
        x = x_new
        iter += 1

        # Check for convergence
        if improvement < tol:
            break

    return x, J_history

# Parameters for Gradient Descent
tol = 1e-6     # Tolerance for convergence
max_iters = 1000  # Maximum number of iterations
initial_x = 5  # Initial value for x

# Different learning rates to experiment with
learning_rates = [0.01, 0.05, 0.1, 0.5, 1.0]

results = []
for alpha in learning_rates:
    # Perform Gradient Descent
    x_final, J_history = gradient_descent(J, dJ, initial_x, alpha, tol, max_iters)

    # Store results: learning rate, final x, number of iterations, final cost
    results.append({
        'learning_rate': alpha,
        'final_x': x_final,
        'num_iterations': len(J_history),
        'final_cost': J_history[-1],
        'J_history': J_history
    })

# Print results
for result in results:
    print(f"Learning Rate: {result['learning_rate']}")
    print(f"Optimized parameter value: {result['final_x']}")
    print(f"Number of iterations: {result['num_iterations']}")
    print(f"Final cost value: {result['final_cost']}")
    print("-" * 50)

# Plotting the cost function values over iterations for each learning rate
plt.figure(figsize=(12, 8))
for result in results:
    plt.plot(result['J_history'], label=f'Learning rate={result["learning_rate"]}')

plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Cost Function Value over Iterations for Different Learning Rates')
plt.legend()
plt.grid(True)
plt.show()
```

Q5) Impact of learning strategies [Optional]

So far, you had a 딲onstant learning rate (alpha). Two popular learning strategies for adapting the learning rate during training are invscaling and adaptive learning rate. You will be implementing these two strategies



The invscaling strategy decreases the learning rate over time according to a schedule. A common schedule is to divide the learning rate by the square root of the iteration number:
alpha = alpha0 / sqrt(t)



where alpha0 is the initial learning rate and t is the iteration number. This strategy is based on the intuition that as the algorithm approaches the minimum, the step size should become smaller to avoid overshooting it.



The adaptive learning rate strategy is a form of learning rate annealing or decay, where the learning rate is reduced to prevent overshooting and oscillations. One common approach is to decrease the learning rate by a factor, such as dividing it by 5 or 10 if overshooting occurs. The change in the sign of the gradient will indicate overshooting during optimization.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function and its gradient
def J(x):
    return x**2 - 4*x + 3

def dJ(x):
    return 2*x - 4

# Gradient Descent with a constant learning rate
def gradient_descent_constant(J, dJ, x, alpha, tol, max_iters):
    J_history = []
    x_history = []
    for iter in range(max_iters):
        # Store history for visualization
        x_history.append(x)
        J_history.append(J(x))

        # Compute the gradient
        grad = dJ(x)

        # Update x
        x_new = x - alpha * grad

        # Check for convergence
        if abs(J(x_new) - J(x)) < tol:
            break

        x = x_new

    return x, J_history, x_history

# Gradient Descent with invscaling learning rate
def gradient_descent_invscaling(J, dJ, x, alpha0, tol, max_iters):
    J_history = []
    x_history = []
    for t in range(1, max_iters + 1):
        alpha = alpha0 / np.sqrt(t)

        # Store history for visualization
        x_history.append(x)
        J_history.append(J(x))

        # Compute the gradient
        grad = dJ(x)

        # Update x
        x_new = x - alpha * grad

        # Check for convergence
        if abs(J(x_new) - J(x)) < tol:
            break

        x = x_new

    return x, J_history, x_history

# Gradient Descent with adaptive learning rate
def gradient_descent_adaptive(J, dJ, x, alpha, tol, max_iters, reduction_factor=5):
    J_history = []
    x_history = []
    prev_sign = np.sign(dJ(x))

    for iter in range(max_iters):
        # Store history for visualization
        x_history.append(x)
        J_history.append(J(x))

        # Compute the gradient
        grad = dJ(x)

        # Update x
        x_new = x - alpha * grad

        # Check if overshooting
        if np.sign(dJ(x_new)) != prev_sign:
            alpha /= reduction_factor

        prev_sign = np.sign(dJ(x_new))

        # Check for convergence
        if abs(J(x_new) - J(x)) < tol:
            break

        x = x_new

    return x, J_history, x_history

# Parameters
initial_x = 5
tol = 1e-6
max_iters = 1000

# Different learning rates and strategies
alpha_constant = 0.1
alpha_invscaling = 0.5
alpha_adaptive = 0.1

# Perform gradient descent with different strategies
results_constant = gradient_descent_constant(J, dJ, initial_x, alpha_constant, tol, max_iters)
results_invscaling = gradient_descent_invscaling(J, dJ, initial_x, alpha_invscaling, tol, max_iters)
results_adaptive = gradient_descent_adaptive(J, dJ, initial_x, alpha_adaptive, tol, max_iters)

# Plotting the results
x_vals = np.linspace(-2, 6, 400)
y_vals = J(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label='Cost Function $J(x)$')

# Plot the paths taken by gradient descent for each strategy
plt.plot(results_constant[2], J(np.array(results_constant[2])), 'o-', label='Constant LR')
plt.plot(results_invscaling[2], J(np.array(results_invscaling[2])), 'o-', label='Invscaling LR')
plt.plot(results_adaptive[2], J(np.array(results_adaptive[2])), 'o-', label='Adaptive LR')

plt.xlabel('$x$')
plt.ylabel('$J(x)$')
plt.title('Gradient Descent Optimization with Different Learning Strategies')
plt.legend()
plt.grid(True)
plt.show()
```
